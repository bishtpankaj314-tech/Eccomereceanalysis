```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df=pd.read_csv("/content/Advanced E-Commerce Dataset for SQL Python or Power BI Projects.csv")
```

```python
df.head()
```

```python
region_visits = df.groupby('Region')['Website_Visits'].sum()

print(region_visits)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=region_visits.index, y=region_visits.values)
plt.title('Website Visits by Region')
plt.xlabel('Region')
plt.ylabel('Total Website Visits')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
total_revenue_by_region = df.groupby('Region')['Revenue'].sum()
print(total_revenue_by_region)
```

```python
avg_discount_by_region = df.groupby('Region')['Discount_Percent'].mean()
print(avg_discount_by_region)
```

```python
correlation = df['Discount_Percent'].corr(df['Revenue'])
print(f"Pearson Correlation between Discount_Percent and Revenue: {correlation:.4f}")
```
